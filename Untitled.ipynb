```python
import numpy as np
import matplotlib.pyplot as plt

# Define the effective potential
def V_eff(r):
    return 0.5 * r**2 + 0.5 / r**2

# Define the total energy
def energy_isoline(E, r):
    return np.sqrt(2 * (E - V_eff(r)))

# Plot the phase portrait
r_values = np.linspace(0.1, 10, 100000)
energy_levels = np.linspace(1, 10, 10)  # Corresponds to E/E0 in the range [E0, 10E0]

plt.figure(figsize=(8,6))

for E in energy_levels:
    r_vals = []
    r_dots = []
    for r in r_values:
        if E > V_eff(r):  # Ensure valid values under the square root
            r_vals.append(r)
            r_dots.append(energy_isoline(E, r))
            r_vals.insert(0, r)
            r_dots.insert(0, -energy_isoline(E, r))
    plt.plot(r_vals, r_dots, label=f'E/E0 = {E:.1f}')

plt.xlabel('r/r0')
plt.xlim(0, 6)
plt.ylabel('dr/dt')
plt.ylim(-6, 6)
plt.title('Phase Portrait for the Fictitious Particle')
plt.legend()
plt.grid(True)
plt.savefig("homework.pdf", format = "pdf")
plt.show()
```
