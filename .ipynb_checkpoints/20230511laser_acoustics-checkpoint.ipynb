# Analyzing laser-sound interactions using laser ablation on a metal plate 

First, we import all of the necessary files and packages that we'll be using for this project.

```python
import numpy
import sys
sys.path.append("../../lhillber/brownian/src")
from time_series import CollectionTDMS as ctdms
import matplotlib.pyplot as plt
```

Here we import the data file, and we process the tdms file into two separate collections.

```python
f = "data/20230511_laser_acoustic.tdms"
laser = ctdms(f)
laser.set_collection("X")
micro = ctdms(f)
micro.set_collection("Y")
```

Now, we can apply some functions to our laser collection. We calibrate it by flipping the sign of our data, and we also create an aggregrate plot to see the average line of our laser data.

```python
laser.apply("calibrate", cal = -1, inplace = True)
laser.apply("detrend", mode = "linear", inplace = True)

# conversion factor 0.00068 V per pascal
fig, ax = plt.subplots(1, 1)

for s in laser.collection[2:]:
    run = s
    run.plot(tmin = 2.825e-4, tmax = 2.975e-4, ax = ax)

laser.aggrigate(collection_slice = slice(2, 11, 1))
laser.agg.plot(tmin = 2.825e-4, tmax = 2.975e-4, ax = ax, c = "k")

plt.title("Voltage vs time reading for the laser aperatus")
plt.xlabel("time (s)")
plt.ylabel("volts (V)")
plt.show()
```

Let's do the same for the microphone data.

```python
micro.apply("detrend", mode = "constant", inplace = True)

fig, ax = plt.subplots(1, 1)

for s in micro.collection[2:]:
    run = s
    run.plot(tmin = 2.8e-4, tmax = 4e-4, ax = ax)

micro.aggrigate(collection_slice = slice(2, 11, 1))
micro.agg.plot(tmin = 2.8e-4, tmax = 4e-4, ax = ax, c = "k")

plt.title("Voltage vs time reading for the microphone aperatus")
plt.xlabel("time (s)")
plt.ylabel("volts (V)")
plt.show()
```

## Power Spectral Density graph

Being able to see this data, let's now look at the Power Spectral Density graphs of this data. I decided to reset the functions we used above, and I filtered out the first two runs of our data (because it gave a null set of data).

```python
laser.apply("restore")
micro.apply("restore")

collection_copies = (laser.collection, micro.collection)
laser.collection = laser.collection[2:]
micro.collection = micro.collection[2:]

laser.apply("lowpass", cutoff = 2e5, inplace = True)
laser.apply("bin_average", Npts = laser.r / (2 * 200000), inplace = True)
micro.apply("lowpass", cutoff = 2e5, inplace = True)
micro.apply("bin_average", Npts = micro.r / (2 * 200000), inplace = True)
laser.average("PSD", taumax = 20e-3)
micro.average("PSD", taumax = 20e-3)

plt.loglog(laser.freq, laser.psd, color = "red")
plt.loglog(micro.freq, micro.psd, color = "yellow")
plt.title("PSD of laser and microphone measurements")
plt.xlabel("frequencies (Hz)")
plt.ylabel("amplitude (g^2/Hz)")
plt.legend(["laser", "microphone"], loc = "best")
plt.xlim([1e3, 2e5])
plt.ylim([1e-14, 1e-8])
plt.show()
laser.collection, micro.collection = collection_copies
```

We know from the microphone's specifications that it's signal acquisition has a transfer function of 0.00068 Volts per pascal. Let's plot the pressure felt by the microphone membrane as a function of time.

```python
micro.apply("restore")
micro.apply("calibrate", cal = (1 / 0.00068), inplace = True)
fig, ax = plt.subplots(1, 1)

for s in micro.collection[2:]:
    run = s
    run.plot(tmin = 2.8e-4, tmax = 4e-4, ax = ax)

micro.aggrigate(collection_slice = slice(2, 11, 1))
micro.agg.plot(tmin = 2.8e-4, tmax = 4e-4, ax = ax, c = "k")

plt.title("pressure vs time on microphone")
plt.xlabel("time (s)")
plt.ylabel("pascal (Pa)")
plt.show()
```

Now we want to come up with a relationship between the pressure data and the voltage reading that we're getting on the laser. We'll start by graphing the aggregate voltages from the two sensors beside each other and seeing the difference.

```python
laser.apply("restore")
micro.apply("restore")
collection_copies = (laser.collection, micro.collection)
laser.collection = laser.collection[:5] + laser.collection[6:]
micro.collection = micro.collection[:5] + laser.collection[6:]


fig, ax = plt.subplots(1, 1)

laser.apply("calibrate", cal=-1, inplace = True)
laser.apply("detrend", mode = "linear", inplace = True)
laser.aggrigate(collection_slice = slice(0, len(laser.collection), 1))
laser.agg.calibrate(cal = 1 / 0.32, inplace = True)
laser.agg.plot(tmin = 2.825e-4, tmax = 4e-4, ax = ax, c = "k")
micro.apply("detrend", mode = "linear", inplace = True)
micro.aggrigate(collection_slice = slice(0, len(micro.collection), 1))
# here we shift the time axis on the laser signal because the microphone is going to take longer to receive the sound pressure wave than the laser based on the slow travel time of the sound wave
# the microphone is somewhere around 5-6 cm from the sound source compared to the laser beam which is only within 1 cm to the source
# and the distance over the speed of sound of the wave creates a delay of about 1.76e-4 s
micro.agg.shift(tau = 1.75e-5, inplace = True)
micro.agg.plot(tmin = 2.825e-4, tmax = 4e-4, ax = ax, c = "r")
plt.legend(["laser", "microphone"], loc = "best")
plt.show()

laser.collection, micro.collection = collection_copies
```
