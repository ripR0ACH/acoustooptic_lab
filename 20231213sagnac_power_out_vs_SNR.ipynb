# Sagnac SNR vs P$_{out}$
On December 12, I started with an input power, $P_{in} = 100.8$ mW $\pm 0.3$ mW, and minimized P$_{out}$ for our Sagnac setup with a maximized P$_{det} = 926.1 \mu$W $\pm 28.3 \mu$W. This is recorded below as `CCW_runs[0]`, because in this run the Sagnac will be perturbed from its best extinction point by CCW rotation of the $\lambda/2$ plate. 

During this run, I also recorded some other key values: SNR$_{mic} = 22.8 \pm 2.5$; and Amp$_{laser} = 1.3\times10^{-2}$V $\pm3.4\times10^{-4}$V. I then changed P$_{det} = 815.6\mu$W $\pm18.7\mu$W and maxed the P$_{out}$ using CW rotation of the $\lambda/2$ plate. This run of measurements is contained within `CCW_runs[1]`. After this set of measurements, I took only one more reliable set of measurements through pushing $P_{det} = 764.9\mu$W $\pm 20.4\mu$W and repeating. This set of measurements is contained within `CCW_runs[2]`.

We then came to the conclusion that the profile of the destructively interferred dark port isn't uniform over all $\phi$ angles for our $\lambda/2$ plate. This meant that it'd be beneficial to take these same measurements for the Sagnac perturbed from its best extinction point by CW rotation of the $\lambda/2$ plate. 

On December 13, I measured the SNR vs P$_{out}$ for rotation in the CW direction. I began by analyzing the beam profile of our dark port. I found that the profile was symmetric enough for the measurements I'd be taking. I then ran into significant problems with the power meter, and after I finally managed to fix it, it seemed that I'd accidentally bumped the split beam detector. This caused the detector to become misaligned, and I spent a while getting correct alignment for this piece. I checked alignment at the end by measuring the individual powers of the split beams entering the detector. I found that I was able to get them to within $\pm 25\mu$W of eachother (resulting in a zeroed split beam detector). This was acceptable for me, so I proceeded to take measurements the same as before.

I recored key values: SNR$_{mic} = 24.7 \pm 2.1$; Amp$_{laser} = 4.0\times10^{-2}$V $\pm4.1\times10^{-3}$V; and Amp$_{mic} = 9.7\times10^{-3}$V $\pm5.2\times10^{-6}$V. I then minimized and measured P$_{out}$ vs. SNR for our Sagnac setup with a maximized P$_{det} = 1.37$mW $\pm 0.03$mW. This is recorded in `CW_runs[0]`. I then adjusted $P_{det} = 945.7\mu$W $\pm14.3\mu$W and took the first series of measurements. These are in `CW_runs[1]`. Lastly, I changed P$_{det} = 483.5\mu$W $\pm 16.6\mu$W and took the last series of measurements. These are recorded in `CW_runs[2]`.

```python
import matplotlib.pyplot as plt
import numpy as np

CCW_Pin = 100.8 #mW
CW_Pin = 110.4 #mW
single_Pin = 108.2 #mW

def sagnac_phase(Pout, Pin):
    return 2 * np.arcsin(np.sqrt(np.array(Pout) / Pin))

CCW_runs = [[[23.8], 
             [1.74]], 
            [[19.1, 21.8, 24.2, 24.5], 
             [2.12, 1.98, 1.86, 1.79]], 
            [[14.5, 15.6, 17.5, 17.5, 20.1, 22.8], 
             [2.30, 2.20, 2.13, 1.94, 1.86, 1.73]]]
err_CCW_runs = [[[2.1], 
                 [0.04]], 
                [[2.2, 1.9, 1.4, 1.6], 
                 [0.04, 0.04, 0.04, 0.04]], 
                [[1.2, 1.7, 1.5, 1.0, 1.7, 1.5], 
                 [0.04, 0.05, 0.05, 0.03, 0.03, 0.04]]]

CW_runs = [[[24.1], 
            [1.37]], 
           [[13.5, 14.2, 14.8, 16.2, 18.3, 19.2, 22.9, 24.1], 
            [2.74, 2.60, 2.51, 2.35, 2.23, 2.12, 2.00, 1.98]], 
           [[4.3, 2.7, 5.9, 6.8, 11.2, 13.1, 13.7, 14.3, 16.1, 17.1, 17.8, 20.0, 24.6, 25.2], 
            [4.30, 4.05, 3.68, 3.53, 3.27, 3.09, 2.98, 2.85, 2.73, 2.55, 2.48, 2.25, 2.13, 2.09]]]
err_CW_runs = [[[2.1], 
                [0.03]], 
               [[2.3, 2.2, 2.6, 2.8, 2.6, 2.1, 2.3, 2.1], 
                [0.03, 0.05, 0.04, 0.04, 0.04, 0.03, 0.02, 0.03]], 
               [[1.2, 1.1, 1.6, 1.7, 1.8, 2.1, 2.1, 1.9, 1.8, 2.3, 2.0, 2.3, 1.8, 1.8], 
                [0.06, 0.06, 0.05, 0.04, 0.05, 0.05, 0.06, 0.06, 0.05, 0.06, 0.06, 0.04, 0.04, 0.06]]]

single_runs = [[0.32, 1.82, 2.15, 2.55, 2.86, 3.44, 4.12, 4.15, 4.48, 4.35, 5.74, 4.95, 6.17, 6.46, 6.10, 5.42, 4.94, 4.05, 3.38], 
               [59.53, 58.42, 58.49, 57.70, 55.91, 55.65, 54.37, 51.47, 48.41, 46.62, 40.28, 38.51, 32.11, 27.81, 20.32, 15.56, 13.48, 10.54, 8.29]]

err_single_runs = [[0.09, 0.12, 0.10, 0.09, 0.13, 0.28, 0.11, 0.41, 0.09, 0.12, 0.09, 0.12, 0.10, 0.08, 0.09, 0.07, 0.07, 0.11, 0.09], 
                   [0.24, 0.32, 0.29, 0.20, 0.27, 0.22, 0.22, 0.30, 0.25, 0.20, 0.17, 0.16, 0.17, 0.12, 0.13, 0.13, 0.11, 0.07, 0.11]]

fig, ax = plt.subplots(1, 1, figsize = (10, 8))
titles = ["Best Run", "Second Run", "Third Run"]
for i in range(len(CCW_runs)):
    ax.set_title(titles[i])
    ax.set_xlabel("P$_{out}$ (mW)")
    ax.set_ylabel("SNR")
    ax.errorbar(sagnac_phase(CCW_runs[i][1], CCW_Pin), CCW_runs[i][0], xerr = 0 *  sagnac_phase(err_CCW_runs[i][1], CCW_Pin), yerr = err_CCW_runs[i][0], label = "CCW run " + str(i), marker = 'o')
    ax.errorbar(sagnac_phase(CW_runs[i][1], CW_Pin), CW_runs[i][0], xerr = 0 * sagnac_phase(err_CW_runs[i][1], CW_Pin), yerr = err_CW_runs[i][0], label = "CW run " + str(i), marker = 'o')
ax.errorbar(sagnac_phase(single_runs[1], single_Pin), single_runs[0], xerr = 0, yerr = err_single_runs[0], label = "single PD", marker = 'o')
ax.legend()
fig.savefig("SNRvsP_out.png")
plt.show()
```
