Importing important packages for our data analysis.

```python
import numpy
import sys
sys.path.append("../../lhillber/brownian/src")
from time_series import CollectionTDMS as ctdms
import matplotlib.pyplot as plt

f = ["data/20230710/Sagnac_500shot_ene_scan/iter_" + str(i) + ".tdms" for i in range(8)]
```

```python
for data in f:
    laser = ctdms(data)
    laser.set_collection("X")
    laser.apply("detrend", mode = "linear", inplace = True)
    micro = ctdms(data)
    micro.set_collection("Y")
    fig, ax = plt.subplots()
    for s in laser.collection:
        s.plot(tmin = 4.5e-4, tmax = 4.8e-4, ax = ax)
    laser.aggrigate(collection_slice = slice(0, len(laser.collection), 1))
    laser.agg.plot(tmin = 4.5e-4, tmax = 4.8e-4, ax = ax, c = "k")
    plt.show()
    fig, ax = plt.subplots()
    for s in micro.collection:
        s.plot(tmin = 3.6e-4, tmax = 4.2e-4, ax = ax)
    micro.aggrigate(collection_slice = slice(0, len(micro.collection), 1))
    micro.agg.plot(tmin = 3.6e-4, tmax = 4.2e-4, ax = ax, c = "k")
    plt.show()
```
