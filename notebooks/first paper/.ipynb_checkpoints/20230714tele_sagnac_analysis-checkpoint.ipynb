```python
import numpy as np
import sys
sys.path.append("../../lhillber/brownian/src")
from time_series import CollectionTDMS as ctdms
import matplotlib.pyplot as plt

f = ["data/20230714/Sagnac_tele_ene_scan_501shots/iter_" + str(i) + ".tdms" for i in range(8)]
for data in f[1:2]:
    laser = ctdms(data)
    micro = ctdms(data)
    laser.set_collection("X")
    micro.set_collection("Y")
    laser.apply("detrend", mode = "linear", inplace = True)
    laser.apply("calibrate", cal = -1, inplace = True)
    fig, ax = plt.subplots()
    for s in laser.collection:
        s.plot(tmin = 4.5e-4, tmax = 4.8e-4, ax = ax)
    laser.aggrigate(collection_slice = slice(0, len(laser.collection), 1))
    laser.agg.plot(tmin = 4.5e-4, tmax = 4.8e-4, ax = ax, c = "k")
    plt.title("Voltage vs time for the laser")
    plt.xlabel("time (s)")
    plt.ylabel("volts (V)")
    plt.show()
    fig, ax = plt.subplots()
    for s in micro.collection:
        s.plot(tmin = 3.6e-4, tmax = 4.2e-4, ax = ax)
    micro.aggrigate(collection_slice = slice(0, len(micro.collection), 1))
    micro.agg.plot(tmin = 3.6e-4, tmax = 4.2e-4, ax = ax, c = "k")
    plt.title("Voltage vs time for the microphone")
    plt.xlabel("time (s)")
    plt.ylabel("volts (V)")
    plt.show()
```

```python
from acoustic_entrainment import mic_response

for data in f[1:2]:
    laser = ctdms(data)
    micro = ctdms(data)
    laser.set_collection("X")
    micro.set_collection("Y")
    snr_l = []
    snr_m = []
    sdev_l = []
    fc = [i for i in range(20000, 1000000, 20000)]
    for freq in fc:
        laser_peak_voltages = []
        micro_peak_voltages = []
        laser_rms = []
        micro_rms = []
        
        laser.apply("detrend", mode = "linear", inplace = True)
        laser.apply("calibrate", cal = -1, inplace = True)
        laser.apply("lowpass", cutoff = freq, inplace = True)
        micro.apply("detrend", mode = "linear", inplace = True)
        micro.apply("lowpass", cutoff = freq, inplace = True)
        micro.apply("correct", response = mic_response, recollect = True)
        
        for shot_laser, shot_micro in zip(laser.collection[1:], micro.collection[1:]):
            laser_peak_voltages.append(max(shot_laser.x))
            laser_rms.append(np.sqrt(np.mean(shot_laser.x ** 2)))
            micro_peak_voltages.append(max(shot_micro.x))
            micro_rms.append(np.sqrt(np.mean(shot_micro.x ** 2)))
        snr_l.append(np.mean(np.array(laser_peak_voltages) / np.array(laser_rms)))
        snr_m.append(np.mean(np.array(micro_peak_voltages) / np.array(micro_rms)))
        laser.set_collection("X")
        micro.set_collection("Y")
    plt.plot(fc, snr_l)
    plt.plot(fc, snr_m)
    plt.title("SNR vs frequency at 13J.")
    plt.xlabel("frequency")
    plt.ylabel("SNR")
    plt.show()
```
