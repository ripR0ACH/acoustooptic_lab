```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
def cos2(x, delta, a, c):
    import numpy as np
    return a * np.power(np.cos(np.radians(x) - delta), 2) + c
def linear(x, a, b):
    return a * x + b
def quad(x, a, b, c):
    import numpy as np
    return a * np.power(x, 2) + b * x + c
def guess(x, a, b, c, d, e, f, g):
    return (a * np.power(x, 3) + b * np.power(x, 2) + c * x + d) / (e * np.power(x, 2) + f * x + g)

pscan_Ene = np.linspace(15, 19, 5) # J
pscan_phi = [np.linspace(0, 155, 32), np.linspace(0, 155, 32), np.linspace(0, 155, 32), np.linspace(0, 160, 33), np.linspace(0, 160, 33)] # deg
pscan_E = [np.array([0.4, 0.9, 1.5, 2.2, 3.1, 3.9, 5.1, 5.9, 6.2, 7.1, 8.0, 8.7, 9.3, 9.9, 10.2, 10.7, 10.8, 10.7, 10.6, 10.3, 10.2, 9.5, 8.8, 7.8, 7.0, 6.0, 5.1, 4.0, 3.1, 2.2, 1.5, 0.8]),
     np.array([0.6, 1.2, 2.1, 3.3, 4.5, 5.9, 7.4, 8.7, 10.3, 11.7, 12.9, 14.0, 15.2, 16.1, 16.6, 17.0, 17.2, 17.5, 17.1, 16.7, 15.9, 14.8, 13.8, 12.3, 10.7, 9.3, 7.7, 6.1, 4.9, 3.5, 2.2, 1.3]),
     np.flip(np.array([1.6, 3.0, 5.1, 8.0, 8.1, 10.2, 12.3, 14.4, 16.3, 18.5, 19.9, 20.9, 22.0, 22.7, 23.3, 23.2, 23.2, 22.7, 21.8, 20.5, 19.4, 17.6, 15.8, 13.9, 11.8, 10.0, 8.1, 6.2, 4.5, 3.0, 1.8, 0.8])),
     np.array([0.95, 2.1, 3.5, 5.3, 7.5, 9.7, 12.2, 14.3, 16.8, 19.1, 21.3, 23.1, 24.8, 26.3, 27.1, 28.1, 28.0, 28.2, 27.4, 26.7, 25.3, 23.7, 21.8, 19.6, 17.3, 14.9, 12.3, 9.95, 7.5, 5.4, 3.5, 1.9, 0.8]),
     np.flip(np.array([0.98, 2.2, 3.95, 6.2, 8.8, 11.4, 14.5, 17.2, 20.3, 23.0, 25.6, 27.5, 30.0, 31.2, 32.4, 33.0, 33.0, 32.9, 32.0, 30.9, 29.1, 27.4, 25.0, 22.3, 19.8, 17.0, 14.4, 11.3, 8.8, 6.4, 4.2, 2.5, 1.2]))
    ] # mJ


fig, ax = plt.subplots(5, 1)
fig.set_size_inches(3, 10)
for i in range(len(pscan_Ene)):
    ax[i].plot(pscan_phi[i], pscan_E[i])
    popt, pcov = curve_fit(cos2, pscan_phi[i], pscan_E[i])
    print("A =", np.abs(popt[1]))
    print("delta =", (180 / np.pi) * (popt[0] % (np.pi / 2)))
    print("const. =", popt[2])
    ax[i].plot(pscan_phi[i], cos2(pscan_phi[i], *popt))

def amp(x, x0, a, b, c):
    return a * (x - x0) * np.abs(x - x0) ** (b - 1) + c

phi = 19
E = [0.47, 0.52, 0.58, 0.65, 0.70, 0.78, 0.86, 0.92, 1.01]
Ene = np.linspace(12.2, 13, 9)
for i in range(len(pscan_E)):
    E.append(pscan_E[i][4])
    Ene = np.r_[Ene, pscan_Ene[i]]
fig, ax = plt.subplots()
ax.plot(Ene, E, "o")
popt, pcov = curve_fit(guess, Ene, E)
ax.plot(np.linspace(10, 20, 100), guess(np.linspace(10, 20, 100), *popt))
print(*popt)
```

```python
from itertools import chain

pscan_Ene = np.linspace(15, 19, 5) # J
pscan_phi = [np.linspace(0, 155, 32), np.linspace(0, 155, 32), np.linspace(0, 155, 32), np.linspace(0, 160, 33), np.linspace(0, 160, 33)] # deg
pscan_E = [np.array([0.4, 0.9, 1.5, 2.2, 3.1, 3.9, 5.1, 5.9, 6.2, 7.1, 8.0, 8.7, 9.3, 9.9, 10.2, 10.7, 10.8, 10.7, 10.6, 10.3, 10.2, 9.5, 8.8, 7.8, 7.0, 6.0, 5.1, 4.0, 3.1, 2.2, 1.5, 0.8]),
     np.array([0.6, 1.2, 2.1, 3.3, 4.5, 5.9, 7.4, 8.7, 10.3, 11.7, 12.9, 14.0, 15.2, 16.1, 16.6, 17.0, 17.2, 17.5, 17.1, 16.7, 15.9, 14.8, 13.8, 12.3, 10.7, 9.3, 7.7, 6.1, 4.9, 3.5, 2.2, 1.3]),
     np.flip(np.array([1.6, 3.0, 5.1, 8.0, 8.1, 10.2, 12.3, 14.4, 16.3, 18.5, 19.9, 20.9, 22.0, 22.7, 23.3, 23.2, 23.2, 22.7, 21.8, 20.5, 19.4, 17.6, 15.8, 13.9, 11.8, 10.0, 8.1, 6.2, 4.5, 3.0, 1.8, 0.8])),
     np.array([0.95, 2.1, 3.5, 5.3, 7.5, 9.7, 12.2, 14.3, 16.8, 19.1, 21.3, 23.1, 24.8, 26.3, 27.1, 28.1, 28.0, 28.2, 27.4, 26.7, 25.3, 23.7, 21.8, 19.6, 17.3, 14.9, 12.3, 9.95, 7.5, 5.4, 3.5, 1.9, 0.8]),
     np.flip(np.array([0.98, 2.2, 3.95, 6.2, 8.8, 11.4, 14.5, 17.2, 20.3, 23.0, 25.6, 27.5, 30.0, 31.2, 32.4, 33.0, 33.0, 32.9, 32.0, 30.9, 29.1, 27.4, 25.0, 22.3, 19.8, 17.0, 14.4, 11.3, 8.8, 6.4, 4.2, 2.5, 1.2]))
    ] # mJ

phi = 19
E = [0.47, 0.52, 0.58, 0.65, 0.70, 0.78, 0.86, 0.92, 1.01]
Ene = np.linspace(12.2, 13, 9)

all_E = np.r_[np.array(list(chain.from_iterable(pscan_E))), E]
all_ene = np.r_[np.array(list(chain.from_iterable([ene * np.ones(len(pscan_Ei)) for ene, pscan_Ei in zip(pscan_Ene, pscan_E)]))), Ene]
all_phi = np.r_[np.array(list(chain.from_iterable(pscan_phi))), [19] * len(E)]
```

```python
def amp(x, x0, a, b, c):
    return a * (x - x0) * np.abs(x - x0) ** (b - 1) + c
def phase(phi, a, phi0, c):
    return a * np.cos(phi - phi0) ** 2 + c
def full(ene_phi, x0, phi0, a, b, c):
    ene, phi = ene_phi
    return amp(ene, x0, a, b, 0) * phase(phi * np.pi / 180, 1, phi0, 0) + c
popt_low, pcov_low = curve_fit(full, (all_ene, all_phi), all_E, sigma = all_E, p0 = (13, 1.4, 5.7, 1.0, 0.5))
popt_high, pcov_high = curve_fit(full, (all_ene, all_phi), all_E, sigma = 1 / all_E, p0 = (13, 1.4, 5.7, 1.0, 0.5))
```

```python
print("Evaluate at 19 J and 84 degrees (low E fit):",  round(full((19, 84), *popt_low), 3))
print("Evaluate at 19 J and 84 degrees (high E fit):",  round(full((19, 84), *popt_high), 3))
```

```python
fig, axs = plt.subplots(2, 1, figsize=(5,7))
for i in range(len(pscan_Ene)):
    color = f"C{i}"
    phi = pscan_phi[i]
    ene = pscan_Ene[i]
    fit_result = [full((ene, ph), *popt_low) for ph in phi]
    axs[0].plot(phi, pscan_E[i], marker="o", ls="none", color=color, label=f"F.L. Ene={int(ene)} J")
    axs[0].plot(phi, fit_result, color=color)

axs[1].plot(Ene, E, marker="o", ls="none", color="k")
axs[1].plot(Ene, [full((ene, 19), *popt_low) for ene in Ene], color="k")

axs[0].legend(loc="upper left")

#axs[0].set_yscale("log")
axs[0].set_ylabel("Pulse energy (mJ)")
axs[0].set_xlabel("Polarizer angle (deg)")

axs[1].text(0.05,0.9, r"Pol. angle = 19 degrees", transform=axs[1].transAxes)
axs[1].set_ylabel("Pulse energy (mJ)")
axs[1].set_xlabel("Flash lamp energy (J)")
print("Low Energy Fit results: ", popt_low)
```
