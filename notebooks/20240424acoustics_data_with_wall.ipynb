```python
import analysis
import numpy as np 
import matplotlib.pyplot as plt

system = analysis.System("sagnac", ["../data/20240422/5cm_right_side_plate_phi8/iter_" + str(x) + ".tdms" for x in range(0, 10)])
```

```python
for i, data in enumerate(system.get_data()):
    fig, ax = plt.subplots(3, 3)
    fig.set_size_inches(12, 15)
    ax = ax.flatten()
    fig.suptitle("Energy " + str(data.ene))
    for j, col in enumerate(data.collection[1:]):
        # ax[j].set_xlim([0.0001594, 0.0001601])
        # ax[j].set_xlim([0.0002365, 0.0002385])
        # ax[j].set_xlim([0.0002265, 0.0002285])
        # ax[j].set_xlim([0.00017102, 0.000172])
        ax[j].plot(col.t, col.x)
    fig.tight_layout()
    fig, ax = plt.subplots()
    s_freq, s_psd = data.average("PSD", tmin = 0.0001594, tmax = 0.000161)
    n_freq, n_psd = data.average("PSD", tmin = 0, tmax = 0.0001594)
    # ax.set_xlim([1, 0.2e8])
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(n_freq, n_psd)
    ax.plot(s_freq, s_psd)
```
