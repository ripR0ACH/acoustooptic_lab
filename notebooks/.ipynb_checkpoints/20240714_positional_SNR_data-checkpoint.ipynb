```python
import sys
sys.path.append("../")
from analysis import System as syst
import matplotlib.pyplot as plt
import numpy as np

linestyle_tuple = [
     (0, (1, 10)),
     (0, (1, 1)),
     (0, (1, 1)),
     (5, (10, 3)),
     (0, (5, 10)),
     (0, (5, 5)),
     (0, (5, 1)),
     (0, (3, 10, 1, 10)),
     (0, (3, 5, 1, 5)),
     (0, (3, 1, 1, 1)),
     (0, (3, 5, 1, 5, 1, 5)),
     (0, (3, 10, 1, 10, 1, 10))]
s = syst("sagnac", ["../../data/20240712/cage_pos_" + str(x + 1) + "/iter_0.tdms" for x in range(8)], snr_resolution = 25, snr_freq_range = [1e4, 1e7], channel = "X", set_data = True)
```

```python
for ind in range(len(s.get_data())):
    s.set_data(ind = ind)
    for i in range(len(s.get_data()[ind].collection) - 1, -1, -1):
        if np.max(np.abs(s.get_data()[ind].collection[i].x)) < 0.05:
            s.get_data()[ind].collection = np.delete(s.get_data()[ind].collection, i)
    for col in range(len(s.get_data()[ind].collection[1:])):
        plt.plot(s.get_data()[ind].collection[col].t, s.get_data()[ind].collection[col].x)
    plt.xlim(0.00018, 0.00022)
    plt.title("Position " + str(ind + 1))
    plt.show()
```

```python
s.reset_snr_vs_freq()
signal = (0.00018, 0.00022)
noise = (0, 0.00018)
for ind in range(len(s.get_data())):
    s.set_data(ind = ind)
    for i in range(len(s.get_data()[ind].collection) - 1, -1, -1):
        if np.max(np.abs(s.get_data()[ind].collection[i].x)) < 0.05:
            s.get_data()[ind].collection = np.delete(s.get_data()[ind].collection, i)
    s.set_snr_vs_freq(s.calc_snr_vs_freq(j = ind, signal = signal, noise = noise, bins = False), ind)
    freq = np.linspace(s.get_snr_freq_range()[0], s.get_snr_freq_range()[1], s.get_snr_resolution())
    plt.plot(freq, s.get_snr_vs_freq()[ind], linestyle = linestyle_tuple[ind], label = "pos " + str(ind + 1))
plt.title("SNR vs Frequency for Positions")
plt.xlabel("Frequency")
plt.ylabel("SNR")
plt.legend(handlelength = 8)
plt.show()
```

```python
col = 2
for ind in range(len(s.get_data())):
    snr = np.array([])
    freq = np.linspace(s.get_snr_freq_range()[0], s.get_snr_freq_range()[1], s.get_snr_resolution())
    for f in freq:
        snr = np.append(snr, s.calc_snr_of_collection_at_cutoff(ind, col, f, signal, noise))
    plt.plot(freq, snr, linestyle = linestyle_tuple[ind], label = "pos " + str(ind + 1))
plt.legend(handlelength = 8)
plt.show()
```

```python
arm1 = np.zeros(25, float)
c1 = 0
arm2 = np.zeros(25, float)
c2 = 0
for ind in range(len(s.get_data())):
    if ind <= 3 and ind != 0:
        arm1 += s.get_snr_vs_freq()[ind]
        c1 += 1
    else:
        arm2 += s.get_snr_vs_freq()[ind]
        c2 += 1
plt.plot(freq, arm1 / c1, label = "First arm")
plt.plot(freq, arm2 / c2, label = "Second arm")
plt.xlabel("Frequency")
plt.ylabel("SNR")
plt.title("Arm 1 vs 2 Comparison")
plt.legend()
plt.show()
```

```python
for ind in range(len(s.get_data())):
    s_freq, s_psd = s.get_data()[ind].average("PSD", tmin = signal[0], tmax = signal[1])
    n_freq, n_psd = s.get_data()[ind].average("PSD", tmin = noise[0], tmax = noise[1])
    plt.loglog(s_freq, s_psd)
    plt.loglog(n_freq, n_psd)
    plt.title("PSD vs Frequency of Position " + str(ind + 1))
    plt.xlabel("Frequency")
    plt.ylabel("PSD")
    plt.show()
```
