# What is going on with this data?
Highpass filter it to see if we can remove the ringing

```python
from scipy import signal as sig

fig, ax = plt.subplots(1, 2, figsize = (20, 10))

sagnac.set_data(ind = 0)
sagnac.set_data(ind = 1)
sagnac.get_data()[0].aggrigate(collection_slice = slice(1, len(sagnac.get_data()[0].collection), 1))
s_freq = [np.array(sagnac.get_data()[0].average("PSD", tmin = 0.00017, tmax = 0.00021)[0]), np.array(sagnac.get_data()[1].average("PSD", tmin = 0.00017, tmax = 0.00021)[0])]
s_psd = [np.array(sagnac.get_data()[0].average("PSD", tmin = 0.00017, tmax = 0.00021)[1]), np.array(sagnac.get_data()[1].average("PSD", tmin = 0.00017, tmax = 0.00021)[1])]
n_freq = [np.array(sagnac.get_data()[0].average("PSD", tmin = 0, tmax = 0.00017)[0]), np.array(sagnac.get_data()[1].average("PSD", tmin = 0, tmax = 0.00017)[0])]
n_psd = [np.array(sagnac.get_data()[0].average("PSD", tmin = 0, tmax = 0.00017)[1]), np.array(sagnac.get_data()[1].average("PSD", tmin = 0, tmax = 0.00017)[1])]
for i in range(len(s_freq)):
    # ax[i].set_xscale("log")
    ax[i].set_yscale("log")
    ax[i].plot(s_freq[i], s_psd[i])
    ax[i].plot(n_freq[i], n_psd[i])
    ax[i].set_xlim([0, 8e6])
plt.show()

fig, ax = plt.subplots(1, 1, figsize = (20, 10))
ax.plot(sagnac.get_data()[0].agg.t, sagnac.get_data()[0].agg.x)
ax.set_xlim([0.00017, 0.0003])
plt.show()

fig, ax = plt.subplots(1, 2, figsize = (20, 10))

sagnac.set_data(ind = 0)
sagnac.set_data(ind = 1)
order = 4
freq = 3e5
for i in range(len(sagnac.get_data())):
    for j in range(len(sagnac.get_data()[i].collection)):
        filt = sig.butter(order, freq, fs = sagnac.get_data()[0].r, btype = "highpass", analog = False, output = "sos")
        sagnac.get_data()[i].collection[j].x = sig.sosfilt(filt, sagnac.get_data()[i].collection[j].x)
filt = sig.butter(1, 7e6, fs = sagnac.get_data()[0].r, btype = "highpass", analog = False, output = "sos")
sagnac.get_data()[0].aggrigate(collection_slice = slice(1, len(sagnac.get_data()[0].collection), 1))
s_freq = [np.array(sagnac.get_data()[0].average("PSD", tmin = 0.00017, tmax = 0.00021)[0]), np.array(sagnac.get_data()[1].average("PSD", tmin = 0.00017, tmax = 0.00021)[0])]
s_psd = [np.array(sagnac.get_data()[0].average("PSD", tmin = 0.00017, tmax = 0.00021)[1]), np.array(sagnac.get_data()[1].average("PSD", tmin = 0.00017, tmax = 0.00021)[1])]
n_freq = [np.array(sagnac.get_data()[0].average("PSD", tmin = 0, tmax = 0.00017)[0]), np.array(sagnac.get_data()[1].average("PSD", tmin = 0, tmax = 0.00017)[0])]
n_psd = [np.array(sagnac.get_data()[0].average("PSD", tmin = 0, tmax = 0.00017)[1]), np.array(sagnac.get_data()[1].average("PSD", tmin = 0, tmax = 0.00017)[1])]
for i in range(len(s_freq)):
    # ax[i].set_xscale("log")
    ax[i].set_yscale("log")
    ax[i].plot(s_freq[i], s_psd[i])
    ax[i].plot(n_freq[i], n_psd[i])
    ax[i].set_xlim([0, 8e6])
plt.show()

fig, ax = plt.subplots(1, 1, figsize = (20, 10))
ax.plot(sagnac.get_data()[0].agg.t, sagnac.get_data()[0].agg.x)
ax.set_xlim([0.00017, 0.0003])
plt.show()
```

In my opinion, this data set needs to be retaken. Too many shots are clipping at +/- 2 V, and the ringing is too prominent of a feature. The former being the more important reason. We're able to see here a much wider BW (expanded by almost 4 times), but this may not be the full extent because of our clipping in the signals. 

After many attempts to reduce the ringing effect using a highpass band filter, it's been mostly unsuccessful.
